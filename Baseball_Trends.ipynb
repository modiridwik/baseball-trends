```python
low = df['AB'].quantile(0.1)
high = df['AB'].quantile(0.9)
med = df['AB'].quantile(0.5)

df['AB'] = np.where(df['AB'] > high, med, df['AB'])
df['AB'] = np.where(df['AB'] < low, med, df['AB'])

low = df['Batting_Average'].quantile(0.1)
high = df['Batting_Average'].quantile(0.9)
med = df['Batting_Average'].quantile(0.5)

df['Batting_Average'] = np.where(df['Batting_Average'] > high, med, df['Batting_Average'])
df['Batting_Average'] = np.where(df['Batting_Average'] < low, med, df['Batting_Average'])

low = df['2B'].quantile(0.1)
high = df['2B'].quantile(0.9)
med = df['2B'].quantile(0.5)

df['2B'] = np.where(df['2B'] > high, med, df['2B'])
df['2B'] = np.where(df['2B'] < low, med, df['2B'])
```
